```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('datasets/reg_data.csv')
df
```

```python
df.dropna(inplace = True)
df
```

```python
y = pd.DataFrame(df['Salary'])
x = pd.DataFrame(df['Age'])

lm = linear_model.LinearRegression()
model_lm = lm.fit(x,y)

model_lm
```

```python
print('The y intercept, b0 is = ',model_lm.intercept_)
print('Thecoef.,b1 is = ',model_lm.coef_)
print('The R^2 is = ',model_lm.score(x,y))
```

```python
# Salary = 349.44580104 + 48.09321584 * Age
```

```python
modelA = sm.OLS.from_formula('Salary ~ Age', data = df)
resultA = modelA.fit()
resultA.summary()
```

```python
print('The salary of a person with age 33 is :', 349.44580104 + 48.09321584 * 33)
```

```python
print('The residual of first two data are:', 1500 - 1936.5219, 1200 - 1936.5219)
```

```python
resiA =resultA.resid
fitValues  = resultA.fittedvalues

resiA
fitValues
```

```python
plt.figure(figsize = (10,8))
ax = sns.regplot(x = 'Age',
                 y = 'Salary',
                 data = df,
                 color = 'darkred',
                 marker = '+',
                 scatter_kws = {'s':200}
                )

ax.set(xlabel = 'Age',
      ylabel = 'Salary')

ax.text(45, 1500, 'Salary = 349.44580104 + 48.09321584 * Age',
        horizontalalignment = 'center', size = 'medium',
        color ='black', weight= 'semibold')

ax.set_title('Regression plot of Age vs Salary')
plt.show()
```

## Conclusion:
#### The Salary is not only depends on Age, but also, some other variables as well.
